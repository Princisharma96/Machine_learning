```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as mlt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
data = df.head(20)
data
```

```python
data.describe(include=['object','bool'])
```

```python
data.sort_values(by='MinTemp',ascending=True)
```

```python
ct = pd.crosstab(data.Rainfall,[data.Evaporation,data.Sunshine],values=data['Rainfall'],aggfunc=np.max)
ct
```

```python
df= pd.crosstab(data.Rainfall,data.Evaporation,values=data['Rainfall'],aggfunc=np.average)
```

```python
df.isnull()
```

```python
import random
df.fillna(random.random())
```

```python
df.dropna()
```

```python
sns.barplot(x = 'Evaporation' , y = 'Sunshine', data = data)
mlt.title('Bar Plot for Evaporation vs Sunshine')
sns.despine()
```

```python
sns.boxplot(x = 'Evaporation' , y = 'Sunshine', data = data)
mlt.title('Box Plot for Evaporation vs Sunshine')
sns.despine()
```

```python
mlt.figure(figsize=(10,8))
sns.countplot(hue='Rainfall', x='Evaporation', data=data)
```

```python
mlt.figure(figsize=(10,8))
corr=df.corr()
sns.heatmap(corr)
```
