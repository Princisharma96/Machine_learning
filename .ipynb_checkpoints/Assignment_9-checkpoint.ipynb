```python
import pandas as pd
import numpy as np
import seaborn as sns
import re
import mpl_toolkits
import matplotlib.pyplot as mlt
from sklearn.preprocessing import *
%matplotlib inline
from sklearn import preprocessing
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')
df = pd.concat([df1, df2])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
dt = df.dropna(subset=["Survived"])
```

```python
dt
```

```python
dt.isnull().sum()
```

```python
dt['Has_Cabin'] = np.where(dt['Cabin'].isnull(), 'yes', 'no')
dt
```

```python
dt['Title']= dt.Name.str.extract(r',\s*([^\.]*)\s*\.', expand=False)
dt
```

```python
dt =dt.fillna(dt.mean())
```

```python
dt.mode()
```

```python
dt["Categorical_age"] = pd.qcut(dt.Age,q= 4,labels = False)
dt["Categorical_fare"] = pd.qcut(dt.Fare,q= 4,labels = False)
dt.head()
```

```python
new = dt.drop(['Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'],axis =1)
new
```

```python
dt= pd.get_dummies(dt)
dt
```

```python
X = dt.drop(["Survived"],axis = 1)
```

```python
Y = dt["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
reg.fit(X_train,Y_train)
```

```python
pred = reg.predict(X_test)
```

```python
result = pd.DataFrame({"Predicted":pred,"Actual":Y_test})
```

```python
result
```

```python
reg.score(X_test,Y_test)*100
```
