```python
data=pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv') 
data.head()
```

```python
sns.distplot(data['Rainfall'])
```

```python
sns.jointplot("MinTemp", "MaxTemp", data, kind='kde')
```

```python
sns.jointplot("MinTemp", "MaxTemp", data, kind='hex')
```

```python
sns.jointplot("Rainfall", "Sunshine", data, kind='reg',color='g')
```

```python
sns.pairplot(data, hue='MaxTemp', size=2.5)
```

```python
sns.distplot(data='MaxTemp',kde=True)
```

```python
sns.kdeplot(data['MaxTemp'],shade=True,color='g')
```

```python
plt.figure(figsize=(10,8))
sns.kdeplot(data['MinTemp'],data['MaxTemp'])
```

```python
sns.joinplot(data['MinTemp'],data['MaxTemp'],kind='hex',color='r') #Hexbin
```

```python
sns.joinplot(data['MinTemp'],data['MaxTemp'],kind='kde',color='r')
```

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x='MinTemp',y='MaxTemp',data= data)
```

```python
plt.figure(figsize=(10,8))
corr=data.corr()
sns.heatmap(corr)
```
