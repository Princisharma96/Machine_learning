```python
import pandas as pd
import numpy as np
import seaborn as sns
import re
import mpl_toolkits
import matplotlib.pyplot as mlt
from sklearn.preprocessing import *
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
%matplotlib inline
from sklearn import preprocessing
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv')
df = pd.concat([df1, df2])
```

```python
a = df.dropna(subset = ['Survived'],inplace=True)
a
```

```python
df.isnull().sum()
```

```python
df['Has_Cabin'] = np.where(df['Cabin'].isnull(), 'yes', 'no')
df
```

```python
df['Title']= df.Name.str.extract(r',\s*([^\.]*)\s*\.', expand=False)
```

```python
df
```

```python
result = df.fillna(df.mean())
result = df.fillna(df.mode)
result.isnull().sum()
```

```python
df['Categorical_age'] = pd.qcut(df['Age'],3 ,labels=False)
df['Categorical_fare'] = pd.qcut(df['Fare'],3 ,labels=False)
df
```

```python
df.drop(['Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'],axis=1)
```
