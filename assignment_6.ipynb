```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as mlt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv')
df = df.head(10)
```

```python
sns.barplot(x = 'Evaporation' , y = 'Sunshine', data = df)
mlt.title('Bar Plot for Evaporation vs Sunshine')
sns.despine()
```

```python
sns.boxplot(x = 'Evaporation' , y = 'Sunshine', data = df)
mlt.title('Box Plot for Evaporation vs Sunshine')
```

```python
sns.pointplot(x = 'MinTemp' , y = 'MaxTemp', data = df)
mlt.title('Point Plot for MinTemp vs MaxTemp')
sns.despine()
```

```python
sns.stripplot(x = 'Evaporation' , y = 'Sunshine', data = df)
mlt.title('Strip Plot for Evaporation vs Sunshine')
sns.despine()
```

```python
sns.jointplot(x = 'MinTemp' , y = 'MaxTemp', data = df)
mlt.title('Joint Plot for MinTemp vs MaxTemp')
sns.despine()
```

```python
sns.swarmplot(x = 'MinTemp' , y = 'MaxTemp', data = df)
mlt.title('Swarm Plot for MinTemp vs MaxTemp')
sns.despine()
```

```python
sns.pointplot(x = 'Humidity9am' , y = 'Humidity3pm', data = df)
mlt.title('Point Plot for Humidity9am vs Humidity3pm')
sns.despine()
```

```python
sns.lmplot(x = 'Humidity9am' , y = 'Humidity3pm' , hue = 'Sunshine', data = df)
mlt.title('lmplot for Humidity9am vs Humidity3pm with hue')
sns.despine()
```

```python
mlt.pie(x = 'Humidity9am', data = df)
mlt.title('Pie Chart for Humidity9am')
sns.despine()
```

```python
sns.distplot(df['Humidity9am'] , kde = False)
mlt.title('distplot for Humidity9am')
sns.despine()
```

```python
sns.pairplot(df , kind = 'reg')
mlt.title('Pair Plot')
sns.despine()
```

```python
sns.factorplot(x = 'Humidity9am' , y = 'Humidity3pm' , hue = 'Sunshine', data = df)
mlt.title('lmplot for Humidity9am vs Humidity3pm with hue')
sns.despine()
```

```python
mlt.pie(x = 'Sunshine', data = df)
mlt.title('Pie Chart for Sunshine')
sns.despine()
```

```python
sns.boxplot(x = 'MaxTemp' , y = 'MinTemp', hue = 'RISK_MM', data = df)
mlt.title('Box Plot for MaxTemp vs MinTemp with hue')
sns.despine()
```

```python
mlt.figure(figsize=(10,8))
corr=df.corr()
sns.heatmap(corr)
```

```python
mlt.figure(figsize=(10,8))
sns.kdeplot(df['MinTemp'],df['MaxTemp'])
```
