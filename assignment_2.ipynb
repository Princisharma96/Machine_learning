```python
# import numpy
import numpy as np
```

```python
# Question 1
z = np.random.randn(10,1)
print(z)
print(z.shape)
print(z.size)
mean = np.mean(z)
print(mean)
```

```python
# Question 2
y = np.random.randn(20,1)
print(y)
print(y.shape)
print(y.size)
var = np.var(y)
print(var)
stnd=np.std(y)
print(stnd)
```

```python
# Question 3
arr1 = np.random.randn(10,20)
arr2 = np.random.randn(20,25)
mul = np.matmul(arr1,arr2)
print(mul)
print(mul.shape)
sum = np.sum(mul)
print(sum)
```

```python
# Question 4
array = []
A = np.random.rand(10,1)
for ele in np.nditer(A):
  ele = 1 / (1 + np.exp(-ele))
  array.append(ele)
arr = np.asarray(array).reshape(10,1)
print(arr)
```
